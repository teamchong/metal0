# Data Analysis with PyAOT

This notebook demonstrates data analysis patterns compiled by PyAOT.

```python
# Setup: Create sample data
data = [10, 20, 30, 40, 50]
print("Data:", data)
```

## Summary Statistics

```python
# Calculate mean
total = 0
count = 0
for value in data:
    total = total + value
    count = count + 1

mean = total / count
print("Mean:", mean)
```

```python
# Find min and max
min_val = data[0]
max_val = data[0]

for value in data:
    if value < min_val:
        min_val = value
    if value > max_val:
        max_val = value

print("Min:", min_val)
print("Max:", max_val)
print("Range:", max_val - min_val)
```

## Data Transformation

```python
# Create new list with transformed values
squared = []
for value in data:
    squared.append(value * value)

print("Squared:", squared)
```

```python
# Filter values
high_values = []
for value in data:
    if value > 25:
        high_values.append(value)

print("High values (>25):", high_values)
```

## Results Summary

```python
# Print final summary
print("Analysis complete!")
print("Processed", count, "data points")
print("Average value:", mean)
```
